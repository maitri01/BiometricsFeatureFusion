```python
class person():
    def __init__(self,main,feature,label):
        self.main=main
        self.feature=feature
        self.template=None
        self.label=label
    def showRep(self):
        plt.figure()
        plt.subplot(1,2,1)
        plt.imshow(self.main)
        plt.subplot(1,2,2)
        plt.imshow(self.feature)
```

```python
class system(person):
    def __init__(self,model):
        self.people=[]
        self.model=model
    def load_data(self,feature_dir,stream2,stream1='rgb'):
        subjects=os.listdir("/content/train")
        subjects.sort()
        pos=0
        count=0
        for i in subjects:
            raw_path=os.path.join("/content/train",i)
            feature_path=os.path.join(feature_dir,i)
            ls_raw = os.listdir(os.path.join(raw_path,stream1))
            ls_raw.sort()
            ls_feature = os.listdir(os.path.join(feature_path,stream2))
            ls_feature.sort()
            # print(os.path.join(os.path.join(raw_path,'rgb'),ls_raw[pos]))
            # print(os.path.join(os.path.join(feature_path,'rgb'),ls_feature[pos]))
            img=cv.imread(os.path.join(os.path.join(raw_path,stream1),ls_raw[pos]),0)
            feature=cv.imread(os.path.join(os.path.join(feature_path,stream2),ls_feature[pos]),0)
            img=img[:128,:128]
            feature=feature[:128,:128]
            
            # print(img.shape)
            self.enroll(main=img,feature=feature,label=count)
            count+=1

        
    def enroll(self,main,feature,label):
        person1=person(main,feature,label)
        person1.label=label
        main=np.reshape(main,(1,128,128,1))
        feature=np.reshape(feature,(1,128,128,1))
        person1.template=self.model.predict((main,feature),batch_size=1,verbose=0,)
        self.people.append(person1)
    def identify(self,main,feature):
        main=np.reshape(main,(1,128,128,1))
        feature=np.reshape(feature,(1,128,128,1))
        temp=self.model.predict((main,feature),verbose=0,)
        val=[]
        temp=np.array(temp).reshape(3,20)
        for person in self.people:
            val.append(np.sum(np.diag(cosine_similarity(temp,np.array(person.template).reshape(3,20)))))
        return val
    def identify_template(self,person):
        temp=person.template
        val=[]
        temp=np.array(temp).reshape(3,20)
        for person in self.people:
            val.append(np.sum(np.diag(cosine_similarity(temp,np.array(person.template).reshape(3,20)))))
        return val
```

```python
class checkSys():

    def __init__(self,sys):
        self.score=[]
        self.system=sys
        self.people=[]
        self.trueIdentity=[]
        self.EER=0
        self.treshEER=0
    def load_data(self,feature_dir,stream2,stream1='rgb'):
        subjects=os.listdir("/content/train")
        subjects.sort()
        pos=50
        count=0
        for i in subjects:
            raw_path=os.path.join("/content/train",i)
            feature_path=os.path.join(feature_dir,i)
            ls_raw = os.listdir(os.path.join(raw_path,stream1))
            ls_raw.sort()
            ls_feature = os.listdir(os.path.join(feature_path,stream2))
            ls_feature.sort()
            # print(os.path.join(os.path.join(raw_path,'rgb'),ls_raw[pos]))
            # print(os.path.join(os.path.join(feature_path,'rgb'),ls_feature[pos]))
            img=cv.imread(os.path.join(os.path.join(raw_path,stream1),ls_raw[pos]),0)
            feature=cv.imread(os.path.join(os.path.join(feature_path,stream2),ls_feature[pos]),0)
            img=img[:128,:128]
            feature=feature[:128,:128]
            
            # print(img.shape)
            self.buildTestData(main=img,feature=feature,label=count)
            img=cv.imread(os.path.join(os.path.join(raw_path,stream1),ls_raw[pos+5]),0)
            feature=cv.imread(os.path.join(os.path.join(feature_path,stream2),ls_feature[pos+5]),0)
            img=img[:128,:128]
            feature=feature[:128,:128]
            
            # print(img.shape)
            self.buildTestData(main=img,feature=feature,label=count)
            count+=1
    def buildTestData(self,main,feature,label):
        person1=person(main,feature,label)
        main=np.reshape(main,(1,128,128,1))
        feature=np.reshape(feature,(1,128,128,1))
        person1.template=self.system.model.predict((main,feature),verbose=0,)
        person1.label=label
        self.people.append(person1)
        self.trueIdentity.append(label)
        self.updateScore()
    def updateScore(self):
        self.score=[]
        for person in self.people:
            val=self.system.identify_template(person)
            self.score.append(val)
    def getScoreCount(self,threshold):
        falseAcc=0
        falseNacc=0
        trueAcc=0
        trueNacc=0
        for i in range(len(self.score)):
            val=np.array(self.score[i])
            val1=np.where(val>threshold,1,0)
            val2=np.where(val<=threshold,1,0)
            if val1[self.trueIdentity[i]]:
                trueAcc+=1
                falseAcc+=(np.sum(val1)-1)
            else:
                falseAcc+=np.sum(val1)
            
            if val2[self.trueIdentity[i]]:
                trueNacc+=1
                falseNacc+=(np.sum(val2)-1)
            else:
                falseNacc+=np.sum(val2)
        return trueAcc,trueNacc,falseAcc,falseNacc
    def showRep(self,start=1,end=3,step=0.001):
        # tresh=np.arange(start,end,step)
        n=40
        tresh=np.arange(1,3,0.001)
        ta=[]
        frr=[]
        far=[]
        tr=[]
        for i in tresh:
            TA,FRR,FAR,TR=self.getScoreCount(i)
            # print(np.max(frr))
            TA/=n
            FRR/=n
            FAR/=n*20-20
            TR/=n*20-20
            if FRR==FAR:
                self.EER=FAR
                self.treshEER=i
            frr.append(FRR)
            far.append(FAR)
            ta.append(TA)
            tr.append(TR)
        # print(np.max(tr))
        # plt.figure(figsize=(10,10))
        plt.plot(tresh, far,label='false acceptance')
        # plt.plot(tresh, ta,label='true acceptance')
        # plt.plot(tresh, tr, '-.',label="true rejection")
        plt.plot(tresh, frr, '-.',label='false rejection')
        plt.xlabel("Threshold")
        plt.ylabel("percent")
        # plt.title('multiple plots')
        plt.legend()
        # plt.show()

```

```python
class mergesys():
    def __init__(self,sys):
        self.systems=sys
        self.scores=np.zeros(np.array(sys[0].score).shape)
        self.trueIdentity=sys[0].trueIdentity
        self.EER=0
        self.treshEER=0
    def getScore(self):
        # self.scores=[]

        for sys in self.systems:
            # print(np.array(sys.score))
            # for person in sys.people:
            #     val=sys.system.identify_template(person)
            #     score.append(val)
            self.scores+=np.array(sys.score)
            # np.sum(self.scores,np.array(sys.score))
    def getScoreCount(self,threshold):
        falseAcc=0
        falseNacc=0
        trueAcc=0
        trueNacc=0
        for i in range(len(self.scores)):
            val=np.array(self.scores[i])
            val1=np.where(val>threshold,1,0)
            val2=np.where(val>threshold,0,1)
            if val1[self.trueIdentity[i]]:
                trueAcc+=1
                falseAcc+=(np.sum(val1)-1)
            else:
                falseAcc+=np.sum(val1)
            if val2[self.trueIdentity[i]]:
                trueNacc+=1
                falseNacc+=(np.sum(val2)-1)
            else:
                falseNacc+=np.sum(val2)
        return trueAcc,trueNacc,falseAcc,falseNacc
    def showRep(self,start=3,end=7,step=0.001):
        tresh=np.arange(start,end,step)
        ta=[]
        frr=[]
        far=[]
        tr=[]
        n=40
        for i in tresh:
            TA,FRR,FAR,TR=self.getScoreCount(i)
            TA/=n
            FRR/=n
            FAR/=n*20-20
            TR/=n*20-20
            if FRR==FAR:
                self.EER=FAR
                self.treshEER=i
            frr.append(FRR)
            far.append(FAR)
            ta.append(TA)
            tr.append(TR)
        # plt.figure(figsize=(10,10))
        plt.plot(tresh, far,label='false acceptance')
        # plt.plot(tresh, ta,label='true acceptance')
        # plt.plot(tresh, tr, '-.',label="true rejection")
        plt.plot(tresh, frr, '-.',label='false rejection')
        plt.xlabel("Threshold")
        plt.ylabel("percent")
        # plt.title('multiple plots')
        plt.legend()
        # plt.show()
```
